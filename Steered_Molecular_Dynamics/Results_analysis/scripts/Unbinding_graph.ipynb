```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "Unbinding_graph.xlsx"
df = pd.read_excel(file_path)

# Rename columns for clarity
df.columns = ['Time_ps', 'ADN_Position_A', 'MOD2_Position_A', 'Force_kcal_mol_A']

# Adjust the force to start from zero
adjusted_force = df['Force_kcal_mol_A'] - df['Force_kcal_mol_A'].iloc[0]

# Create the figure
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot ADN and MOD2 positions (left Y-axis)
ax1.plot(df['Time_ps'], df['ADN_Position_A'], label='ADN', color='blue')
ax1.plot(df['Time_ps'], df['MOD2_Position_A'], label='MOD2', color='green')
ax1.set_xlabel('Time (ps)', fontsize=14)
ax1.set_ylabel('Position ()', fontsize=14)
ax1.tick_params(axis='both', labelsize=12)
ax1.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Plot adjusted force (right Y-axis)
ax2 = ax1.twinx()
ax2.plot(df['Time_ps'], adjusted_force, label='Force Applied', color='red', linestyle='--')
ax2.set_ylabel('Force Applied (kcal/mol)', fontsize=14)
ax2.tick_params(axis='y', labelsize=12)

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, fontsize=12, loc='upper left')

# Title and layout
plt.title('Ligand Position and Applied Force Over Time', fontsize=16)
fig.tight_layout()
plt.show()
```
