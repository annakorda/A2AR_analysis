```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the two exported PMF files
pmf1 = pd.read_csv("pmf_output_adn.dat", sep=r"\s+", names=["Z", "PMF_ADN"])
pmf2 = pd.read_csv("pmf_output_mod2.dat", sep=r"\s+", names=["Z", "PMF_MOD2"])

# Create the figure
plt.figure(figsize=(8, 5))
plt.plot(pmf1["Z"], pmf1["PMF_ADN"], label="ADN", color="blue")
plt.plot(pmf2["Z"], pmf2["PMF_MOD2"], label="MOD2", color="green")
plt.xlabel("Position ()", fontsize=13)
plt.ylabel("Work / PMF (kcal/mol)", fontsize=13)
plt.title("PMF Comparison", fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, linestyle="--", linewidth=0.5, alpha=0.7)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```
